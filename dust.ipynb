```python
import netCDF4 as nc
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
from rasterio import features
from matplotlib.colors import LogNorm
from scipy import stats

# Path to the NetCDF file
file_path = r'C:\Users\Subhadeep Maishal\Music\JUP\Breathe\pollutants_BC_OC_DUST.nc'

# Open the NetCDF file
nc_file = nc.Dataset(file_path, 'r')

# Read the variables
dust_mass = nc_file.variables['DUSMASS25'][:]
bc_mass = nc_file.variables['BCSMASS'][:]
ocs_mass = nc_file.variables['OCSMASS'][:]
latitude = nc_file.variables['lat'][:]
longitude = nc_file.variables['lon'][:]
longitude[longitude > 180] -= 360  # Convert longitudes to range from -180 to 180

# Function to calculate linear trend and 95% confidence interval
def calculate_trend(data):
    num_time_steps = data.shape[0]
    trend = np.zeros_like(data[0])
    ci = np.zeros_like(data[0])
    p_values = np.zeros_like(data[0])

    for i in range(data.shape[1]):
        for j in range(data.shape[2]):
            y = data[:, i, j]
            x = np.arange(num_time_steps)
            slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
            trend[i, j] = slope
            p_values[i, j] = p_value
            ci[i, j] = std_err * 1.96  # 95% confidence interval

    return trend, ci, p_values

# Calculate trend for each variable
dust_mass_trend, dust_mass_ci, dust_mass_p_values = calculate_trend(dust_mass)
bc_mass_trend, bc_mass_ci, bc_mass_p_values = calculate_trend(bc_mass)
ocs_mass_trend, ocs_mass_ci, ocs_mass_p_values = calculate_trend(ocs_mass)

# Close the NetCDF file
nc_file.close()

# Plotting
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Plot DUSMASS25 trend
im1 = axs[0].imshow(dust_mass_trend, cmap='jet', origin='lower', extent=[longitude.min(), longitude.max(), latitude.min(), latitude.max()])
axs[0].set_title('PM$_{2.5}$ Trend')
axs[0].set_xlabel('Longitude')
axs[0].set_ylabel('Latitude')
cbar1 = fig.colorbar(im1, ax=axs[0], label='PM$_{2.5}$ (log scale) [kg m$^{-3}$]', fraction=0.02)  # Set colorbar height
cbar1.ax.yaxis.set_tick_params(width=0)  # Set colorbar ticks to be invisible

# Plot BCSMASS trend
im2 = axs[1].imshow(bc_mass_trend, cmap='jet', origin='lower', extent=[longitude.min(), longitude.max(), latitude.min(), latitude.max()])
axs[1].set_title('Black Carbon Trend')
axs[1].set_xlabel('Longitude')
axs[1].set_ylabel('Latitude')
cbar2 = fig.colorbar(im2, ax=axs[1], label='Black carbon (log scale) [kg m$^{-3}$]', fraction=0.02)  # Set colorbar height
cbar2.ax.yaxis.set_tick_params(width=0)  # Set colorbar ticks to be invisible

# Plot OCSMASS trend
im3 = axs[2].imshow(ocs_mass_trend, cmap='jet', origin='lower', extent=[longitude.min(), longitude.max(), latitude.min(), latitude.max()])
axs[2].set_title('Organic Carbon Trend')
axs[2].set_xlabel('Longitude')
axs[2].set_ylabel('Latitude')
cbar3 = fig.colorbar(im3, ax=axs[2], label='Organic carbon (log scale) [kg m$^{-3}$]', fraction=0.02)  # Set colorbar height
cbar3.ax.yaxis.set_tick_params(width=0)  # Set colorbar ticks to be invisible

# Overlay continents on each subplot
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
for ax in axs:
    world.plot(ax=ax, facecolor='none', edgecolor='black')

# Set longitude limits
for ax in axs:
    ax.set_xlim(longitude.min(), longitude.max())

plt.tight_layout()
plt.show()
```
